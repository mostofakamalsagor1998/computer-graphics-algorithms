```python
def cyrus_beck_algorithm(x_min,x_max,y_min,y_max,x0,y0,x1,y1):
    if x0==x1 and y0==y1:
        print("Line is degenerated, so clip as a point!")
    else:
        info={"Ni":[],"D":[],"Ni*D":[],"t":[],"PE/PL":[],"tE":[],"tL":[]}

        D=x1-x0,y1-y0
        tE=0
        tL=1

        N=-1,0
        N_D=N[0]*D[0]
        if N_D!=0:
            tLeft=-(x0-x_min)/(x1-x0)
            info["Ni"]+=[N]
            info["D"]+=[D]
            info["Ni*D"]+=[N_D]
            info["t"]+=[tLeft]

            if N_D<0:#Angle>90*
                info["PE/PL"]+=["PE"]
            else:#Angle<90*
                info["PE/PL"]+=["PL"]

            if info["PE/PL"][0]=="PE":
                tE=max(tE,tLeft)
            if info["PE/PL"][0]=="PL":
                tL=min(tL,tLeft)

            info["tE"]+=[tE]
            info["tL"]+=[tL]

        N=1,0
        N_D=N[0]*D[0]
        if N_D!=0:
            tRight=-(x0-x_max)/(x1-x0)
            info["Ni"]+=[N]
            info["D"]+=[D]
            info["Ni*D"]+=[N_D]
            info["t"]+=[tRight]

            if N_D<0:#Angle>90*
                info["PE/PL"]+=["PE"]
            else:#Angle<90*
                info["PE/PL"]+=["PL"]

            if info["PE/PL"][1]=="PE":
                tE=max(tE,tRight)
            if info["PE/PL"][1]=="PL":
                tL=min(tL,tRight)

            info["tE"]+=[tE]
            info["tL"]+=[tL]

        N=0,-1
        N_D=N[1]*D[1]
        if N_D!=0:
            tBottom=-(y0-y_min)/(y1-y0)
            info["Ni"]+=[N]
            info["D"]+=[D]
            info["Ni*D"]+=[N_D]
            info["t"]+=[tBottom]

            if N_D<0:#Angle>90*
                info["PE/PL"]+=["PE"]
            else:#Angle<90*
                info["PE/PL"]+=["PL"]

            if info["PE/PL"][2]=="PE":
                tE=max(tE,tBottom)
            if info["PE/PL"][2]=="PL":
                tL=min(tL,tBottom)

            info["tE"]+=[tE]
            info["tL"]+=[tL]

        N=0,1
        N_D=N[1]*D[1]
        if N_D!=0:
            tTop=-(y0-y_max)/(y1-y0)
            info["Ni"]+=[N]
            info["D"]+=[D]
            info["Ni*D"]+=[N_D]
            info["t"]+=[tTop]

            if N_D<0:#Angle>90*
                info["PE/PL"]+=["PE"]
            else:#Angle<90*
                info["PE/PL"]+=["PL"]

            if info["PE/PL"][3]=="PE":
                tE=max(tE,tTop)
            if info["PE/PL"][3]=="PL":
                tL=min(tL,tTop)

            info["tE"]+=[tE]
            info["tL"]+=[tL]
        
        if tE>tL:
            print("\n>>>NULL : Line is comepletely outside!")
        else:
            print("\n>>>True clip intersection: ")
            print(f"P(tE) = P({tE}) = {tE*D[0]+x0,tE*D[1]+y0}")
            print(f"P(tL) = P({tL}) = {tL*D[0]+x0,tL*D[1]+y0}")
            
        df = pd.DataFrame(info).style.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
        return df.set_properties(**{'text-align': 'center'}).hide_index()
```
