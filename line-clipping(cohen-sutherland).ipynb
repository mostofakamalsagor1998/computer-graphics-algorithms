<h1>Line Clipping (Cohen Sutherland)</h1>

```python
import pandas as pd
```

```python
def calculateOutcode(minimum_x,maximum_x,minimum_y,maximum_y,x,y):
    
    # x_min<=x<=x_max
    # y_min<=y<=y_max

    #left boundary
    x_min=minimum_x# MAKE CHANGES HERE...
    #right boundary
    x_max=maximum_x# MAKE CHANGES HERE...
    #bottom boundary
    y_min=minimum_y# MAKE CHANGES HERE...
    #top boundary
    y_max=maximum_y# MAKE CHANGES HERE...
    
    # region_code={"above":0,"below":0,"right":0,"left":0}
    bit=[0,0,0,0]
    
    # Divides co-ordinate system into 9 regions...
    # window_left_top_corner="1001"
    # window_middle_top="1000"
    # window_right_top_corner="1010"
    # window_left="0001"
    # window="0000"
    # window_right="0010"
    # window_left_bottom_corner="0101"
    # window_middle_bottom="0100"
    # window_right_bottom_corner="0110"
    

    if x<x_min:
        bit[3]=1
    if x>x_max:
        bit[2]=1
    if y<y_min:
        bit[1]=1
    if y>y_max:
        bit[0]=1
         
    return bit

def findIntersectionPointOfLine(x_min,x_max,y_min,y_max,xt,yt,m,oc):
    msg=""
    if oc[3]==1:
        #left boundary
        x=x_min
        y=yt+m*(x_min-xt)
        msg="[Left Boundary or Edge] "+msg
        
    if oc[2]==1:
        #right boundary
        x=x_max
        y=yt+m*(x_max-xt)
        msg="[Right Boundary or Edge] "+msg
    
    if oc[1]:
        #bottom boundary
        y=y_min
        x=xt+1/m*(y_min-yt)
        msg="[Bottom Boundary or Edge] "+msg
        
    if oc[0]:
        #top boundary 
        y=y_max
        x=xt+1/m*(y_max-yt)
        msg="[Top Boundary or Edge]"+msg
    
    return x,y,msg
    
    
def cohen_sutherland_algorithm(minimum_x,maximum_x,minimum_y,maximum_y,x1,y1,x2,y2):
    
    oc1=calculateOutcode(minimum_x,maximum_x,minimum_y,maximum_y,x1,y1)
    oc2=calculateOutcode(minimum_x,maximum_x,minimum_y,maximum_y,x2,y2)
    outcode_1=["".join(map(str,oc1))]
    outcode_2=["".join(map(str,oc2))]
    
    m=(y2-y1)/(x2-x1)
    
    m_str=None
    
    #Extra processing...
    a=y2-y1
    b=x2-x1
    a_f=[]
    b_f=[]
    c=[]
    for i in range(1,abs(a)+1):
        if abs(a)%i==0:
            a_f+=[i]
    for j in range(1,abs(b)+1):
        if abs(b)%j==0:
            b_f+=[j]  
    for e1 in a_f:
        if e1 in b_f:
            c+=[e1]
    if m<0:
        m_str=str((y2-y1)//max(c))+"/"+str((x2-x1)//max(c))
    else:
        m_str=str(abs(y2-y1)//max(c))+"/"+str(abs(x2-x1)//max(c))
    
    #...
    
    m_val1=[]
    m_val2=[]
    
    A=[]#Nearest integer...
    B=[]#Nearest integer...
    A.append((round(x1),round(y1)))
    B.append((round(x2),round(y2)))
    
    RA=[]
    RB=[]
    RA.append((round(x1,2),round(y1,2)))
    RB.append((round(x2,2),round(y2,2)))
    
    itr=0
    
    msg1=[]
    msg2=[]
    
    while True:
        if oc1==oc2==[0,0,0,0]:
            msg1+=["[Completely Inside]"]
            msg2+=["[Completely Inside]"]
            m_val1+=[m_str]
            m_val2+=[m_str]
            break
        
        elif [oc1[0] and oc2[0],oc1[1] and oc2[1],oc1[2] and oc2[2],oc1[3] and oc2[3]]!=[0,0,0,0]:
            msg1+=["[Completely Outside]"]
            msg2+=["[Completely Outside]"]
            m_val1+=[m_str]
            m_val2+=[m_str]
            break
            
        else:
            if oc1!=[0,0,0,0]:
                x1,y1,msg=findIntersectionPointOfLine(minimum_x,maximum_x,minimum_y,maximum_y,x1,y1,m,oc1)
                oc1=calculateOutcode(minimum_x,maximum_x,minimum_y,maximum_y,x1,y1)
                msg1+=[msg]
                msg2+=["[Partially Inside]"]
                m_val1+=[ m_str]
                m_val2+=["---"]
            elif oc2!=[0,0,0,0]:
                x2,y2,msg=findIntersectionPointOfLine(minimum_x,maximum_x,minimum_y,maximum_y,x2,y2,m,oc2)
                oc2=calculateOutcode(minimum_x,maximum_x,minimum_y,maximum_y,x2,y2)
                msg1+=["[Completely Inside]"]
                msg2+=[msg]
                m_val1+=["---"]
                m_val2+=[m_str]
        A.append((round(x1),round(y1)))
        B.append((round(x2),round(y2)))
        RA.append((round(x1,2),round(y1,2)))
        RB.append((round(x2,2),round(y2,2)))
        outcode_1+=["".join(map(str,oc1))]
        outcode_2+=["".join(map(str,oc2))]

    df=pd.DataFrame({"Step":[x for x in range(1,len(outcode_1)+1)],"Real(x1,y1)":RA,"Rounded(x1,y1)":A,"m1":m_val1,"Outcode 1":outcode_1,"Intersection 1":msg1,"Real(x2,y2)":RB,"Rounded(x2,y2)":B,"m2":m_val2,"Outcode 2":outcode_2,"Intersection 2":msg2})
    df = df.style.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
    return df.set_properties(**{'text-align': 'center'}).hide_index()
```
