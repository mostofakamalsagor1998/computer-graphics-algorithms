<h1>Mid Point Line & 8-Way Symmetry</h1>

```python
import math as m
import pandas as pd
```

```python
def straightLineEQtoEndPoints(gradient,cuts_in_y_axis):
    # set values...
    m=gradient
    c=cuts_in_y_axis

    # calculate first point...
    x1=0
    y1=round(m*x1+c)

    # calculate second point...
    y2=0
    x2=round((y2-c)/m)
    return x1,y1,x2,y2#starting point : (x1,y1) and ending point : (x2,y2)
```

<h4>Equilateral Triangle</h4>

```python
def equilateralTriangle_A(x1,y1,side):
    half=side//2
    x2=x1+side
    y2=y1
    x3=x1+half
    y3=round(y1+m.sqrt(side**2-half**2))
    return x2,y2,x3,y3

def equilateralTriangle_B(x2,y2,side):
    half=side//2
    x1=x2-side
    y1=y2
    x3=x2-half
    y3=round(y2+m.sqrt(side**2-half**2))
    return x1,y1,x3,y3

def equilateralTriangle_C(x3,y3,side):
    half=side//2
    x1=x3-side
    y1=round(y3-m.sqrt(side**2-half**2))
    x2=x3+side
    y2=round(y3-m.sqrt(side**2-half**2))
    return x1,y1,x2,y2

def printCoordinates(A,B,C):
    print("Original Co-ordinate: ")
    print("A",A)
    print("B",B)
    print("C",C)
```

```python
def zone0conversion(x1,y1,x2,y2):
    
    #calculate dx,dy...
    dx=x2-x1
    dy=y2-y1
    
    #check for zone...
    if dx>=0 and dy>=0 and abs(dy)>=abs(dx):#zone 1: 110
        return [1,y1,x1,y2,x2]
    elif dx<0 and dy>=0 and abs(dy)>=abs(dx):#zone 2: 010
        return [2,y1,-x1,y2,-x2]
    elif dx<0 and dy>=0 and abs(dx)>=abs(dy):#zone 3: 011
        return [3,-x1,y1,-x2,y2]
    elif dx<0 and dy<0 and abs(dx)>=abs(dy):#zone 4: 001
        return [4,-x1,-y1,-x2,-y2]
    elif dx<0 and dy<0 and abs(dy)>=abs(dx):#zone 5: 000
        return [5,-y1,-x1,-y2,-x2]
    elif dx>=0 and dy<0 and abs(dy)>=abs(dx):#zone 6: 100
        return [6,-y1,x1,-y2,x2]
    elif dx>=0 and dy<0 and abs(dx)>=abs(dy):#zone 7: 101
        return [7,x1,-y1,x2,-y2]
    elif dx>=0 and dy>=0 and abs(dx)>=abs(dy):#zone 6: 111
        return 0,x1,y1,x2,y2
        
```

```python
def originalZoneConversion(z,x,y):
    if z==1:
        return y,x
    elif z==2:
        return y,-x
    elif z==3:
        return -x,y
    elif z==4:
        return -x,-y
    elif z==5:
        return -y,-x
    elif z==6:
        return -y,x
    elif z==7:
        return x,-y
    else:
        return x,y
```

```python
def midPointLine(X,Y):
    #Input parameters
    x1,y1,x2,y2=X[0],X[1],Y[0],Y[1]
    #Original-zone-representation
    x,y=[],[]
    x.append(x1)
    y.append(y1)
    
    zone=[]
    
    d_update=[]#Next step
    
    move=[]#NE(+1,-1),E(+1)
    
    itr=0
    
    #After zone conversion...
    z,x1,y1,x2,y2=zone0conversion(X[0],X[1],Y[0],Y[1])
    coordinates=[(x1,y1)]    
    dx=x2-x1
    dy=y2-y1

    d=[2*dy-dx]
    delta_NE=2*dy-2*dx
    delta_E=2*dy

    print(f"dx = {dx} , dy = {dy}")
    print(f"d = {d[0]} , delta_NE = {delta_NE} , delta_E = {delta_E}")
        
    print(f"\nOrizinal zone : {z}\n")
  
    while x1<=x2:
        if d[itr]>0:
            d_update+=[d[itr]+delta_NE]
            d+=[d_update[itr]]
            x1+=1
            y1+=1
            move+=["NE"]#NE-->NORTH-EAST 
        elif d[itr]<=0:
            d_update+=[d[itr]+delta_E]
            d+=[d_update[itr]]
            x1+=1
            move+=["E"] #E-->EAST
        coordinates+=[(x1,y1)]
        #Back to original zone...
        xt,yt=originalZoneConversion(z,x1,y1)
        x.append(xt)
        y.append(yt)
        
        itr+=1
    d.pop()

    df={"SL":[x for x in range(1,itr+1)],
        "Move":move[:itr],"Zone-0":coordinates[:itr],
        "d":d[:itr],"d-update":d_update[:itr],
        "x":x[:itr],"y":y[:itr]}
    df=pd.DataFrame(df)
    df = df.style.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
    
#     File saving...
#     path="D:\\CSE\CSE423\\17301106\\Theory Practice Problem\\mid-point-line_8-way-symmetry\\code-generated-output\\"
#     filename=input("Enter filename : ")
#     df.set_properties(**{'text-align': 'center'}).to_excel(f"{path}{filename}.xlsx")

    return df.set_properties(**{'text-align': 'center'}).hide_index()
```
